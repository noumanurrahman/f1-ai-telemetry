```python
def get_telemetry(session, driver):
    laps = session.laps.pick_drivers(driver)
    telemetry = []
    for lap in laps.iterlaps():
        telemetry.append(lap[1].telemetry)
    return telemetry
```

```python
get_telemetry(session, "HAM")
```

```python
drivers = session.laps['Driver'].unique()
all_telemetry = {}
for driver in drivers:
    telemetry = get_telemetry(session, driver)
    all_telemetry[driver] = telemetry

all_telemetry
```
